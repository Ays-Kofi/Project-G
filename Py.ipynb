```python
import numpy as np
import pandas as pd
import matplotlib. pyplot as plt
import csv
import seaborn as sns
```

```python
Treenumber = 'datasets/primary.csv' 
```

```python
Treenumber = pd.read_csv('datasets/primary.csv')
```

```python
Treenumber.head()
```

```python
description=Treenumber.describe().T
print(description)
```

```python
plt.figure(figsize=(12, 8))

plt.bar(Treenumber['Year'], Treenumber['Eastern Samar'], color=['#fbf8cc', '#ffcfd2', '#cfbaf0', '#a3c4f3', '#90dbf4', '#8eecf5', '#98f5e1', '#b9fbc0'])

plt.title("Land Cover of the Philippines in year 2011 to 2019")
plt.xlabel('Year', fontsize=14)
plt.ylabel('Eastern Samar', fontsize=14)
plt.grid(True)

plt.show()
```

```python
description = Treenumber.describe()
plt.figure(figsize=(12, 8))

description.loc['mean'].plot(kind='bar')

plt.xlabel('Columns')
plt.ylabel('Mean')
plt.title('Mean Values of Columns')

plt.show()
```

```python
description = Treenumber.describe()

# Extract the statistics you want to visualize
statistics = description.loc[['count', 'mean', 'std', 'min', '25%', '50%']]

# Transpose the DataFrame for easier plotting
statistics = statistics.T

# Adjust figure size
plt.figure(figsize=(12, 8))

# Plotting the line graph
statistics.plot(kind='line', marker='o')

# Adding labels and title
plt.xlabel('Columns')
plt.ylabel('Value')
plt.title('Descriptive Statistics')

# Adjust font size
plt.xticks(range(len(statistics.index)), statistics.index, fontsize=8)
plt.xticks(rotation=90)
# Display the plot
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))                                                                         

plt.title('Tree Cover of Phillipines in year of 2012 to 2019')       

for year in Treenumber:                                                                      
    plt.plot(Treenumber[year], label=year, marker = 'h')

                                           
plt.xlabel('Year', fontsize=11)                                                               

plt.legend(loc='upper center', bbox_to_anchor=(1.19, .8))                                          
plt.show()      
```
